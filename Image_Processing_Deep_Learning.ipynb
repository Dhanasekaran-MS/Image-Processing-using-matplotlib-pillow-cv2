```python
import urllib.request
import os

# URL of the image
url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRIB7JceX1qFCrLzau0lzsR9N8m2G7HJsKIBw&s.jpg'
new_filename = 'dog.jpg'

# Download the image and save it as a temporary file
urllib.request.urlretrieve(url, new_filename)

# Rename the temporary file to the new filename
# os.rename(temp_filename, )

print(f"Image downloaded and saved as {new_filename}")
```
